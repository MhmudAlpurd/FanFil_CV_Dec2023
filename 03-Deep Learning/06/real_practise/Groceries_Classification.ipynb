```python
img = []
tag = []
imgsize = 150

def importing_data(label, data_dir):
    middle_path = 'freiburg_groceries_dataset_sample'
    data_dir = os.path.join(middle_path, data_dir)
    print(data_dir)
    for image in tqdm(os.listdir(data_dir)):
        path = os.path.join(data_dir,image )
        print(path)
        image = cv2.imread(path, cv2.IMREAD_COLOR)
        image = cv2.resize(image, (imgsize, imgsize))
        
        img.append(np.array(image))
        tag.append(str(label))
        
```

```python
importing_data('beans', beans)
importing_data('cake', cake)
importing_data('cereal', cereal)
```
